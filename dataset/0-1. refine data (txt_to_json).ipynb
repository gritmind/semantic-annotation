### Insert data to dictionaries (before making json)

```python
def find_ann(ann_num, i, char, list_sentence):
    ann_array = []
    if list_sentence[i]=='$' and list_sentence[i+1]==ann_num:  
        for j, char2 in enumerate(list_sentence):
            if j<=i+1:
                continue
                
#             if list_sentence[j] != list_sentence[-1]:
            if list_sentence[j]==ann_num:
                if list_sentence[j+1]=='$': 
                    break
            if list_sentence[j]==' ':
                if list_sentence[j+1]=='$': 
                    break
                    
            ann_array.append(char2) # until end of sentence
        ann_array.append("*") # is there multiple annotations
    return ann_array

def refine_ann_array(ann_array):
    temp_array = []
    temp_array = ''.join(ann_array)
    temp_array = temp_array.split("*")
    temp_array = filter(None, temp_array)
    

    if temp_array != []:  
        new = []
        for i, word in enumerate(temp_array):
            word_list = list(word) # divide

            for j, char in enumerate(word_list):
                if word_list[-1] in ('.', ',', ':'):
                    del word_list[-1]
                new.append(''.join(word_list)) # merge
                break
        temp_array = new
        
    if len(temp_array)==1:
        #print temp_array
        temp_array = ''.join(temp_array) # ['abc'] -> 'abc'
        
        if(len(list(temp_array))!=0):
            #print temp_array
            char_array = list(temp_array)
            if char_array[0] == ' ':
                del char_array[0]
            #print char_array[-1]
            if char_array[-1] == ' ':
                del char_array[-1]
            temp_array = ''.join(char_array)
        
    return temp_array
```

```python
#def data_to_json(data)
dict_array = []

for sentence in dirty_data:
    
    list_sentence = list(sentence)
     
    component_array = []
    refinement_of_component_array = []
    action_array = []
    argument_of_action_array = []
    condition_array = []
    priority_array = []
    motivation_array = []
    role_array = []
    object_array = []
    refinement_of_object_array = []
    sub_action_array = []
    sub_argument_of_action_array = []
    sub_priority_array = []
    sub_role_array = []
    sub_object_array = []
    sub_refinement_of_object_array = []

    dict_format = {
        '0': '', '1___component': [], '2___refinement_of_component': [], '3___action': [], 
        '4___argument_of_action': [], '5___condition': [], '6___priority': [], '7___motivation': [], '8___role': [], '9___object': [],
        'a___refinement_of_object': [], 'b___sub_action': [], 'c___sub_argument_of_action': [], 'd___sub_priority': [],
        'e___sub_role': [], 'f___sub_object': [], 'g___sub_refinement_of_object': []
    }
    
    # Find each annotation from the sentence
    for i, char in enumerate(list_sentence):
        
        component_array += find_ann('1', i, char, list_sentence)   
        refinement_of_component_array += find_ann('2', i, char, list_sentence)
        action_array += find_ann('3', i, char, list_sentence)
        argument_of_action_array += find_ann('4', i, char, list_sentence)
        condition_array += find_ann('5', i, char, list_sentence)
        priority_array += find_ann('6', i, char, list_sentence)
        motivation_array += find_ann('7', i, char, list_sentence)
        role_array += find_ann('8', i, char, list_sentence)
        object_array += find_ann('9', i, char, list_sentence)
        refinement_of_object_array += find_ann('a', i, char, list_sentence)
        sub_action_array += find_ann('b', i, char, list_sentence)
        sub_argument_of_action_array += find_ann('c', i, char, list_sentence)
        sub_priority_array += find_ann('d', i, char, list_sentence)
        sub_role_array += find_ann('e', i, char, list_sentence)
        sub_object_array += find_ann('f', i, char, list_sentence)
        sub_refinement_of_object_array += find_ann('g', i, char, list_sentence)

            
    # Clean the dirty sentence
    for i, char in enumerate(list_sentence):
            
        if(list_sentence[i] == '$'):
            if(list_sentence[i+1] == '1' or '2' or '3' or '4' or '5' or '6' or '7' or '8' or '9' or 'a' or 'b' or 'c' or 'd' or 'e' or 'f' or 'g'):
                if list_sentence[i+1] != ' ':
                    #print list_sentence[i+1]
                    del(list_sentence[i], list_sentence[i])
    for i, char in enumerate(list_sentence):    
        if(i!=0):
            if(list_sentence[i] == '$'):
                if(list_sentence[i-1] == '1' or '2' or '3' or '4' or '5' or '6' or '7' or '8' or '9' or 'a' or 'b' or 'c' or 'd' or 'e' or 'f' or 'g'):    
                    #print list_sentence[i-1], list_sentence[i] 
                    del(list_sentence[i-1], list_sentence[i-1])
    
    
    # Add annotations to dictionary        
    dict_format['0'] = refine_ann_array(list_sentence)
    dict_format['1___component'] = refine_ann_array(component_array)
    dict_format['2___refinement_of_component'] = refine_ann_array(refinement_of_component_array)
    dict_format['3___action'] = refine_ann_array(action_array)
    dict_format['4___argument_of_action'] = refine_ann_array(argument_of_action_array)
    dict_format['5___condition'] = refine_ann_array(condition_array)
    dict_format['6___priority'] = refine_ann_array(priority_array)
    dict_format['7___motivation'] = refine_ann_array(motivation_array)
    dict_format['8___role'] = refine_ann_array(role_array)
    dict_format['9___object'] = refine_ann_array(object_array)
    dict_format['a___refinement_of_object'] = refine_ann_array(refinement_of_object_array)
    dict_format['b___sub_action'] = refine_ann_array(sub_action_array)
    dict_format['c___sub_argument_of_action'] = refine_ann_array(sub_argument_of_action_array)
    dict_format['d___sub_priority'] = refine_ann_array(sub_priority_array)
    dict_format['e___sub_role'] = refine_ann_array(sub_role_array)
    dict_format['f___sub_object'] = refine_ann_array(sub_object_array)
    dict_format['g___sub_refinement_of_object'] = refine_ann_array(sub_refinement_of_object_array)
    
    
    # Add sentence info to dictionary array
    dict_array.append(dict_format)
    #return json_data
```
